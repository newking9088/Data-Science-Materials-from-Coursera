```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\newki\OneDrive\Desktop\GDP data_till2019.csv',skiprows = 4)
df.head();
```

```python
df = df[(df['Country Name'] == 'Sri Lanka') |(df['Country Name'] == 'Maldives') | (df['Country Name'] == 'Bangladesh') | (df['Country Name'] == 'Afghanistan') | (df['Country Name'] == 'Bhutan') | (df['Country Name'] == 'Nepal') | (df['Country Name'] == 'India') | (df['Country Name'] == 'Pakistan')] 
```

```python
df.reset_index()
df = df.drop(columns = ['Country Code','Indicator Name','Indicator Code'])
cols = df.columns.drop('Country Name')
```

```python
pd.to_numeric(cols)
```

```python
df = df.set_index('Country Name')
```

```python
df = df.T;
df.head()
```

```python
df
```

```python
df.plot(colormap='viridis', figsize=(12,5), lw=4, kind = 'line');
plt.gca().set_xlabel('Years')
plt.ylabel('GDP in US Dollars $')
plt.title('GDP change of SAARC Countries 1960- 2019')
plt.savefig(r'C:\Users\newki\OneDrive\Desktop\GDP.png')
plt.show()
```
